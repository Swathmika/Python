```python
import numpy as np
import matplotlib.pyplot as plt
import math
from mpl_toolkits.mplot3d import Axes3D


pi = math.pi
h = 8.77  #Heat constant 
Te = 300.15  #Environment/Surrounding Temperature
thetha = 163.70  #Angle at the center of the breast
r = 0.07  #Radius of Breast
arclen = r*thetha*(pi/180)  #Formula for arc length calculation

a = np.zeros(200) 
k = -arclen/2
count = 0

while k <= (arclen/2):
    a[count] = k
    k += 0.001
    count += 1

#Defining the points and range on the breast surface as an array

print(a)
```

```python
#To get the temperature values by varying tumor parameters
 
d = np.zeros(5) #depth of tumor
i = 0.01
count = 0
while i <= 0.018:
    d[count] = i
    i += 0.002
    count += 1
    
R = np.zeros(5) # radius of tumour
k=0.001
count = 0
while k <= 0.005:
    R[count] = k
    k += 0.001
    count += 1
    
Q = np.zeros(5) #variable heat intensity
l=0.02
count = 0
while l <= 0.1:
    Q[count] = l
    l += 0.02
    count += 1

T = np.zeros((5, 200)) 
#This is the shape of the "empty" matrix to be filled later

i = 1
for i in range(1, 5):
    for j in range(1, 200):
        for k in range(1,5):
            for l in range(1,5):
                T = (Te + (Q[l]/(4*h*(((d[i]+R[k])**2)+(a[j]**2)))))-273.15  #temperature at the surface calculation
                print(T)
#Now we have a matrix with the calculated temperatures for each point
```
